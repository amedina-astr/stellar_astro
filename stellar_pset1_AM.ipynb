**Problem 1b:**

- Install spectrum synthesis program Korg
- Make an example synthetic stellar spectrum with your choice of:
    - effective temperature
    - surface gravity
    - metallicity
- Make a second spectrum, but at a different effective temperature. What is different between the two spectra? 

Running the example

```julia
using Korg
using Plots

wls, flux, continuum = synth(
    Teff=5000,
    logg=4.32,
    m_H=-1.1,
    C=-0.5,
    linelist=Korg.get_GALAH_DR3_linelist(),
    wavelengths=(5850, 5900)
)

plot(wls, flux,
    xlabel="Wavelength []",
    ylabel="Flux [erg s cm]",
    title="Synthetic Spectrum",
    legend=false,
    linewidth=1.5)
```

Running my own with the same parameters but different effective temperature

```julia
wls2, flux2, continuum2 = synth(
    Teff=6500,
    logg=4.32,
    m_H=-1.1,
    C=-0.5,
    linelist=Korg.get_GALAH_DR3_linelist(),
    wavelengths=(5850, 5900)
)

plot(wls2, flux2,
    xlabel="Wavelength []",
    ylabel="Flux [erg s cm]",
    title="Synthetic Spectrum",
    legend=false,
    linewidth=1.5)
```

For sanity - to test what I am observing.

```julia
wls3, flux3, continuum3 = synth(
    Teff=4500,
    logg=4.32,
    m_H=-1.1,
    C=-0.5,
    linelist=Korg.get_GALAH_DR3_linelist(),
    wavelengths=(5850, 5900)
)

plot(wls3, flux3,
    xlabel="Wavelength []",
    ylabel="Flux [erg s cm]",
    title="Synthetic Spectrum",
    legend=false,
    linewidth=1.5)
```

So it appears that when all the parameters stay the same, except the effective temperature changes, the peaks(or rather dips) in flux occur around the same wavlengths.

But, at higher temperatures, they are noticably more shallow and more narrow comapred to lower temperatures. After consulting the literature, this occurs becasue hotter stars have more highly ionized atoms, leading to weaker line absorption, and flux is overall hgiher due to increased thermal emission from the hotter star.
